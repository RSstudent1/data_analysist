```python
# Trzeci przykład
#Liczba_decyzji_uznanie_za_obywatela_polskiego

uznanie = pd.read_csv('Liczba_decyzji_uznanie_za_obywatela_polskiego.csv')
#print(uznanie)
uznanie = uznanie[uznanie['Liczba osób'] >= 10].reset_index()[['Obywatelstwo','Liczba decyzji','Liczba osób']]
print(uznanie)
```

```python
fig, ax = plt.subplots(figsize=(25,10))
#ax.plot(uznanie['Obywatelstwo'], uznanie['Liczba decyzji'], uznanie['Liczba osób'])
#plt.bar_label(ax.containers[0])
#plt.subplot(fig)

x = np.arange(len(uznanie['Obywatelstwo']))  # the label locations
width = 0.2  # the width of the bars

test11 = uznanie['Liczba decyzji']
test22 = uznanie['Liczba osób']
#print(test11)

rects1 = ax.bar(x - width/2, test11, width, label='ilosc uznan')
rects2 = ax.bar(x + width/2, test22, width, label='ilosc osob')

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.set_xticks(x, uznanie['Obywatelstwo'])
ax.legend()

#plt.figure(figsize=(20,40))
#uznanie.plot.bar() 
plt.savefig('uznanieObywatelstwa.pdf', dpi=240, format='pdf', bbox_inches='tight')
```

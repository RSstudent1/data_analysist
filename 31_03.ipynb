```python
oceny = pd.read_excel('dane1.xlsx',sheet_name='Arkusz4',header=1, names=['Przedmiot', 'Ocena'])
print(oceny) 
```

```python
oceny.groupby('Przedmiot').count() 
```

```python
oceny.groupby('Przedmiot').mean()
#lub
oceny.groupby('Przedmiot').Ocena.mean() 
```

```python
oceny.groupby(['Przedmiot','Ocena']).Ocena.count()
```

```python
oceny.groupby('Przedmiot').agg({'Ocena':['min','max']}) 
```

```python
oceny.groupby('Przedmiot').agg(Najniższa_ocena=('Ocena','min'),Najwyższa_ocena=('Ocena','max'))
```

```python
oceny[oceny['Ocena'] >
2.0].groupby('Przedmiot').agg(Najniższa_ocena=('Ocena','min'),Najwyższa_ocena=('Ocena','max'))
```
